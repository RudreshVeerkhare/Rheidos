# Kung Fu Panda Interactive Demo
Follow the cells below to interactively control Panda3D from the notebook.


```python
from kung_fu_panda.engine import Engine
from kung_fu_panda.views import AxesView, MeshSurfaceView, MeshWireframeView
from kung_fu_panda.resources import cube
from kung_fu_panda.controllers import (
    ExitController,
    FpvCameraController,
    ToggleViewController,
)

eng = Engine(window_title="Kung Fu Panda — Interactive", interactive=True)
eng
```

Instantiate simple helpers. The window opens immediately because the engine auto-starts in interactive mode.


```python
eng.add_view(AxesView(axis_length=2.0))
```

```python
primitive = cube(size=1.5)
eng.add_view(MeshSurfaceView(primitive.mesh, name="surface"))
eng.add_view(MeshWireframeView(primitive.mesh, name="wireframe"))
eng.enable_view("wireframe", False)
```

```python
eng.add_controller(FpvCameraController())
eng.add_controller(ToggleViewController(eng, groups=[["surface"], ["wireframe"]], key="space"))
eng.add_controller(ExitController(eng))
```

Use WASD + mouse to fly, SPACE to toggle wireframe, ESC to close the window.
